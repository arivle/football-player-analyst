# Player Analyst

## Overview

In this project, I tried to make a suggestion for Arsenal, whom are going to be Arsenal midfielders next season. The data was obtained from FBRef. 

## Preparation

First, we have to import mandatory libraries like pandas, numpy, matplotlib, and radar chart by soccerplots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

## Preparing the Data

As we want to know all players' attributes, I collected two data, first the general attributes like goals and assists, and the defensive attributes

```python
df_player_general = pd.read_csv("./data/player-data-general.csv")
df_player_general
```

```python
df_player_defend = pd.read_csv("./data/player-data-defend.csv")
df_player_defend
```

```python
df_player = pd.concat([df_player_general, df_player_defend], axis=1)
df_player = df_player.loc[:, ~df_player.columns.duplicated()].copy()
df_player
```

```python
df_player_standard = df_player[['Rk', 'Player', 'Squad', 'Pos', 'Age', 'Min', '90s', 
                                'PrgC', 'PrgP', 'PrgR','AstPer90', 'G-PKPer90', 
                                'G+A-PKPer90', 'xAGPer90', 'npxGPer90', 'npxG+xAGPer90', 
                                'Tkl', 'Tkl%', 'Lost', 'Blocks', 'Int', 'Clr', 'Err']].copy() 
df_player_standard
```

## Evaluate Current Performance

For evaluating current performance, we groupped the midfielders by their team. I just need to compare the performance of top four teams. For avoiding the outliers, I just used the data of more than 1,000 minutes played players. For Arsenal, I made a dataframe which contained the possible midfielders would play for next season (2023/2023).

### Arsenal Midfielder

```python
df_arsenalMid = df_player_standard[(df_player_standard["Squad"]=="Arsenal") & (df_player_standard["Pos"].str.contains("MF"))]
```

```python
df_arsenalMid = df_player_standard[(df_player_standard["Squad"]=="Arsenal") & (df_player_standard["Pos"].str.contains("MF"))]
df_arsenalMid = df_arsenalMid[df_arsenalMid["Min"] >=1000]
df_arsenalMid.sort_values(by='Min', ascending=False)
```

```python
df_arsenalSUM = df_arsenalMid.cumsum()
df_arsenalSUM = df_arsenalSUM[-1:].drop(["Rk", "Player", "Squad", "Pos", "Age", "Min", "90s"], axis=1)
df_arsenalSUM.insert(loc=0, column = 'Team', value= 'Arsenal')
df_arsenalSUM
```

## Attacking and Defensive Performance

```python
def ranges(df, params):
    ranges = []
    first_values = []
    second_values = []
    team1 = str(df['Team'].iloc[0])
    team2 = str(df['Team'].iloc[1])
    
    for i in params:
        a = min(df[params][i])
        a = round((a - (a*.2)),2) 
        b = max(df[params][i])
        b = round((b + (b*.2)),2)
        ranges.append((a,b))
    
    return ranges
```

## PL Midfielder Profiles

Understandable that the Arsenal's problem is depth. They only had three midfielders were used regularly by Arteta as their outstanding performance and the benchers had not good enough yet. In this section, I tried to find out the most wanted midfielders' profile. There are plenty attributes, but my consideration were the younger ages and perfectly balanced performance.  

```python
df_Mid = df_player_standard[(df_player_standard["Min"]>=1000) & (df_player_standard["Pos"].str.contains("MF"))]

df_Mid['PrgC'] = round((df_Mid['PrgC']/df_Mid['90s']), 2)
df_Mid['PrgP'] = round((df_Mid['PrgP']/df_Mid['90s']), 2)
df_Mid['PrgR'] = round((df_Mid['PrgR']/df_Mid['90s']), 2)

df_Mid
```

```python
def find_weight(df, params):
    weights = []
        
    for i in params:
        w = round((df[params][i].std() / df[params][i].mean()), 2)
        weights.append(w)
    
    return weights

defensive_params = ['Tkl', 'Lost', 'Blocks', 'Int', 'Clr', 'Err']
attacking_params = ['PrgC', 'PrgP', 'PrgR', 'AstPer90', 'G-PKPer90', 'xAGPer90', 'npxGPer90']
defensive_weights = find_weight(df_Mid, defensive_params)
attacking_weights = find_weight(df_Mid, attacking_params)
```

```python
defensive_weights
```

```python
attacking_weights
```

```python
df_Mid['DefensivePoint'] = ((df_Mid['Tkl']*0.39) - (df_Mid['Lost'])) + ((df_Mid['Blocks']) + (df_Mid['Int']) + (df_Mid['Clr']) - (df_Mid['Err']))
df_Mid['AttackingPoint'] = (df_Mid['PrgC'] * 0.58) + (df_Mid['PrgP'] * 0.45) + (df_Mid['PrgR'] * 0.67) + (df_Mid['AstPer90'] * 0.86) + (df_Mid['G-PKPer90'] * 0.94) + (df_Mid['xAGPer90'] * 0.62) + (df_Mid['npxGPer90'] * 0.71)
```

### Defensive Midfielders

Like the name. Defensive midfielder should be decent at defending. So, in this part, I took all of midfielders that have defensive points higher than the attacking points.

```python
df_dm = df_Mid_Rank[(df_Mid_Rank['AttackingPoint'] < df_Mid_Rank['DefensivePoint']) & (df_Mid_Rank["Pos"]=="MF") & (df_Mid_Rank["Min"] > 0.5)].sort_values(by='Total', ascending = False)
df_dm['Rank'] = (df_dm['Rank'].rank(method='first', ascending = True).astype('int'))
df_dm
```

```python
df_dm_barplot = df_dm.drop(['Rk', 'Squad', 'Pos', '90s'], axis=1)
df_dm_barplot = df_dm_barplot.sort_values(by='Rank', ascending =False)
df_dm_barplot[['Age', 'Min', 'DefensivePoint', 'AttackingPoint']] = df_dm_barplot[['Age', 'Min', 'DefensivePoint', 'AttackingPoint']].apply(lambda x: x*100)

fields = ['Age', 'Min', 'DefensivePoint', 'AttackingPoint']
labels = ['Age', 'Stamina', 'Defending', 'Attacking']
colors = ['#1D2F6F', '#FFBF00', '#5A8D03', '#B62203']

fig, ax = plt.subplots(figsize=(20,15))

left = len(df_dm_barplot) * [0]
for i, player in enumerate(fields):
    bars = plt.barh(df_dm_barplot['Player'], df_dm_barplot[player], left=left, color=colors[i])
    ax.bar_label(bars, color = 'white', label_type='center', padding = -1)
    left = left + df_dm_barplot[player]
    
# title, legend, labels
plt.title('Top Premier League DMs\n', loc='left')
plt.legend(labels, bbox_to_anchor=([0.55, 1, 0, 0]), ncol=4, frameon=False)
plt.xlabel('Potential')# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# adjust limits and draw grid lines
plt.ylim(-0.5, ax.get_yticks()[-1] + 0.5)
ax.set_axisbelow(True)
ax.xaxis.grid(color='gray', linestyle='dashed')
plt.show()
```
