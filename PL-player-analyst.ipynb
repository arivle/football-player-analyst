```python
#get parameters
params = list(df_arsenal_att.columns)
params = params[1:]
params
```

```python
def comparison(df, params):
    ranges = []
    first_values = []
    second_values = []
    team1 = str(df['Team'].iloc[0])
    team2 = str(df['Team'].iloc[1])
    
    for i in params:
        a = min(df[params][i])
        a = a - (a*.25)
        
        b = max(df[params][i])
        b = b + (b*.25)
        
        ranges.append((a,b))
    
    for i in range(len(df['Team'])):
        if df['Team'].iloc[i] == team1:
            first_values = df.iloc[i, 1:].values.tolist()
        if df['Team'].iloc[i] == team2:
            second_values = df.iloc[i, 1:].values.tolist()
    
    return ranges, first_values, second_values 
```

```python
df_Mid['PrgC'] = round((df_Mid['PrgC']/df_Mid['90s']), 2)
df_Mid['PrgP'] = round((df_Mid['PrgP']/df_Mid['90s']), 2)
df_Mid['PrgR'] = round((df_Mid['PrgR']/df_Mid['90s']), 2)
```

```python
df_Mid
```

```python
def find_weight(df, params):
    weights = []
        
    for i in params:
        w = round((df[params][i].std() / df[params][i].mean()), 2)
        weights.append(w)
    
    return weights
```

```python
defensive_params = ['Tkl', 'Lost', 'Blocks', 'Int', 'Clr', 'Err']
attacking_params = ['PrgC', 'PrgP', 'PrgR', 'AstPer90', 'G-PKPer90', 'xAGPer90', 'npxGPer90']
```

```python
defensive_weights = find_weight(df_Mid, defensive_params)
defensive_weights
```

```python
attacking_weights = find_weight(df_Mid, attacking_params)
attacking_weights
```

```python
df_Mid['DefensivePoint'] = ((df_Mid['Tkl']*0.39) - (df_Mid['Lost'])) + ((df_Mid['Blocks']) + (df_Mid['Int']) + (df_Mid['Clr']) - (df_Mid['Err']))
```

```python
df_Mid.sort_values(by='DefensivePoint', ascending=False)
```

```python
df_Mid['AttackingPoint'] = (df_Mid['PrgC'] * 0.58) + (df_Mid['PrgP'] * 0.45) + (df_Mid['PrgR'] * 0.67) + (df_Mid['AstPer90'] * 0.86) + (df_Mid['G-PKPer90'] * 0.94) + (df_Mid['xAGPer90'] * 0.62) + (df_Mid['npxGPer90'] * 0.71)
```

```python
df_Mid.sort_values(by='AttackingPoint', ascending=False)
```

```python
df_Mid['Difference'] = df_Mid['AttackingPoint'] - df_Mid['DefensivePoint']
df_Mid[(df_Mid['Difference'] >=0)].sort_values(by='Difference', ascending=True)
```

```python
df_Mid_Rank = df_Mid.drop(['PrgC', 'PrgP', 'PrgR', 'AstPer90', 'G-PKPer90', 
                                     'G+A-PKPer90', 'xAGPer90', 'npxGPer90', 'npxG+xAGPer90','Tkl', 'Tkl%', 
                                     'Lost', 'Blocks', 'Int', 'Clr', 'Err', 'Difference'],axis=1)
df_Mid_Rank
```

```python
df_Mid_Rank['DefensivePoint'] = round((df_Mid_Rank['DefensivePoint'].rank(pct=True)),2)
df_Mid_Rank['Min'] = round((df_Mid_Rank['Min'].rank(pct=True)),2)
df_Mid_Rank['AttackingPoint'] = round((df_Mid_Rank['AttackingPoint'].rank(pct=True)),2)
```

```python
df_Mid_Rank['Total'] = df_Mid_Rank['Min'] + df_Mid_Rank['DefensivePoint'] + df_Mid_Rank['AttackingPoint']
df_Mid_Rank['Rank'] = round((df_Mid_Rank['Total'].rank(pct=True)),2)
df_Mid_Rank = df_Mid_Rank.sort_values(by='Rank', ascending=False)
df_Mid_Rank['Rank'] = (df_Mid_Rank['Total'].rank(method='first', ascending = False).astype('int'))
df_Mid_Rank
```
